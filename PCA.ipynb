```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from numpy import linalg as LA
```

```python
iris = datasets.load_iris() 
X = iris.data
y = iris.target
names = iris.target_names
```

```python
x1,x2,x3,x4=[],[],[],[]
for i in range(len(X)):
    x1.append(X[i][0]),x2.append(X[i][1]),x3.append(X[i][2]),x4.append(X[i][3])
print(x1)
```

```python
x1m=sum(x1)/len(x1)
x2m=sum(x2)/len(x2)
x3m=sum(x3)/len(x3)
x4m=sum(x4)/len(x4)
print(x1m,x2m,x3m,x4m)
```

```python
xx1,xx2,xx3,xx4=[],[],[],[]
for i in range(len(x1)):
    t=x1[i]-x1m
    xx1.append(t)
    t=x2[i]-x2m
    xx2.append(t)
    t=x3[i]-x3m
    xx3.append(t)
    t=x4[i]-x4m
    xx4.append(t)
#print(xx1,xx2,xx3,xx4)
```

```python
A=np.array([xx1,xx2,xx3,xx4])
print(A)
```

```python
B=A.T               
print(B)
```

```python
CovM=np.dot(A,B)
CovM=(1/(len(X)-1))*CovM
print(CovM)
```

```python
ev, e_v = LA.eigh(CovM)     #Eigenvalue   and Eigenvectors
print('Eigenvalue: ',ev)
print('Eigenvectors: \n',e_v)
```

```python
p=ev
mi1=p.argmax(axis=0)
p2=np.delete(p,mi1)
mi2=p.argmax(axis=0)
if mi2==mi1:
    mi2-=1
elif mi2>mi1:
    mi2+=1
print(mi1,mi2)
print(e_v[mi1])
print(e_v[mi2])
```

```python
pc1=[]
pc2=[]
for i in range(len(X)):
    pc1.append(np.dot(B[i],e_v[mi1]))
    pc2.append(np.dot(B[i],e_v[mi2]))
#print(pc1)
```

```python
total=sum(ev)
print(ev[mi1]*100/total ,'%')
print(ev[mi2]*100/total ,'%')
```

```python
final=np.array([pc1,pc2])
print(final)
```

```python
plt.scatter(pc1,pc2, color = 'green') 
plt.show()
```
